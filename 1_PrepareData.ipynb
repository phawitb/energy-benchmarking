```python
import pandas as pd
```

```python
df = pd.read_csv("data/raw_data_phawit_revised.csv",header=1)
df.shape
```

```python
df
```

```python
df = df.fillna(0)
df
```

```python
df = df.drop(columns=['id'])
df
```

```python
# Converting all columns in the DataFrame to numeric data types, where possible
df = df.replace({',': ''}, regex=True).apply(pd.to_numeric, errors='coerce')
df
```

```python
# Deleting rows where 'total_energy_use_kBtu' equals 0
df = df[df['total_energy_use_kBtu'] != 0]
df
```

```python
df.to_csv('data/raw_data_phawit_revised_init.csv', index=False)
```

```python
# Initialize a dictionary to store the counts
count_dict = {}

# Iterate through each column
for column in df.columns:
#     null_count = df[column].isnull().sum()
    zero_count = (df[column] == 0).sum()
    count_dict[column] = {'Null Values': null_count, 'Zero Values': zero_count}

# Display the counts
i = 1
for column, counts in count_dict.items():
    print(f'Column {i}: {column},Zero Values: {counts["Zero Values"]} from {df.shape[0]}')
    i += 1
```
